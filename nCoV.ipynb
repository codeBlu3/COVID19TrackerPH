```python
import pandas as pd
import numpy as np
from geotext import GeoText
import re

df = pd.read_csv('rappler_scraping1.csv')
confirmed = pd.read_csv('rappler_scraping_confirmed_nCov.csv')

df = df.append(confirmed).reset_index()
```

```python
df.head()
```

```python
df.info()
```

```python
df['ncov'] = df['title'].str.find(r'coronavirus')
```

```python
df_ncov = df[df['ncov']>0]
```

```python
df_ncov.info()
```

```python
df_ncov.columns
```

```python
df1 = df_ncov[['source_id', 'date', 'category', 'title', 'author', 'text']].reset_index()
df1['places'] = df1['title'].apply(lambda x: GeoText(x))
df1['cities'] = df1['places'].apply(lambda x: list(x.cities))
df1.head()
```

```python
near_num = r"(?:[a-zA-Z'-]+[^a-zA-Z'-]+){0,3}[0-9](?:[^a-zA-Z'-]+[a-zA-Z'-]+){0,3}"

df1['near_num'] = df['text'].apply(lambda x: re.findall(near_num, x))
```

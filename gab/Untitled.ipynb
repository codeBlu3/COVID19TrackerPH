```python
def scrape_wiki():

    # Wikipedia link of interest

    pages = "https://en.wikipedia.org/wiki/List_of_shopping_malls_in_the_Philippines"

    # Use beautiful soup to scrape and put into a dataframe list

    reqs = requests.get(pages, headers={'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64)'})
    soup = BeautifulSoup(reqs.content, 'html.parser')
    description = soup.find_all("table", class_="wikitable sortable")
    df = pd.read_html(str(description))
    
    pages = 'https://en.wikipedia.org/wiki/List_of_shopping_malls_in_Metro_Manila'
    # Use beautiful soup to scrape and put into a dataframe list

    reqs = requests.get(pages, headers={'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64)'})
    soup = BeautifulSoup(reqs.content, 'html.parser')
    description = soup.find_all("table", class_="wikitable sortable")
    df2 = pd.read_html(str(description))

     # Combine all tables and finalize fields
    df = pd.concat(df[0:18]).reset_index(drop=True)
    df2 = pd.concat(df2[0:8]).reset_index(drop=True)
    df_final = pd.concat([df, df2]).reset_index(drop=True)
    df_final = df_final[['Name','Developer', 'Floor area (m²)', 'Building', 'City/municipality', 'Image', 'Location', 'Province', 
                     'Remarks', 'Retail Space (m2)', 'Year opened']]

    df_final.to_csv('wiki_raw.csv', index=False)
    df.to_csv('wiki_manila.csv', index = False)
    df2.to_csv('wiki_non_manila.csv', index = False)

    return df_final
```

```python
import geopandas as gpd
from shapely.geometry import Point
```

```python
def psgc_integ(mall_list):
    bgy_df = gpd.read_file('Barangays.shp')
    psgc_df = pd.DataFrame(columns = ['Reg_Name', 'Pro_Name', 'Mun_Name', 'Bgy_Name', 'Bgy_Code'])
    mall_list['points'] = [Point(xy) for xy in zip(mall_list.LONG, mall_list.LAT)]

    for points in mall_list['points']:
        try:
            psgc_df = psgc_df.append(bgy_df[bgy_df['geometry'].contains(points)][['Reg_Name', 'Pro_Name', 'Mun_Name', 'Bgy_Name','Bgy_Code']])

        except IndexError:
            psgc_df = psgc_df.append(pd.Series(), ignore_index=True)

    psgc_df = psgc_df.reset_index(drop = True)

    mall_list = mall_list.merge(psgc_df, left_index=True, right_index=True)
    mall_list = mall_list.rename(columns={'Reg_Name':'REGION', 'Pro_Name':'PROVINCE', 'Mun_Name':'ÇITY', 'Bgy_Name':'BRGY', 'Bgy_Code':'PSGC'})
    mall_list['ID'] = mall_list.index
    return mall_list
```

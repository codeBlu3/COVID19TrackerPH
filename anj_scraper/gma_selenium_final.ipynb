```python
import urllib
import pandas as pd
import time

from selenium import webdriver 
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC 
from selenium.common.exceptions import TimeoutException

import datetime

def get_philstar():
    # Download chromedriver and place it in the same path

    # Launch Chrome (or any browser)
    option = webdriver.ChromeOptions()

    # Create new Instance of Chrome
    driver = webdriver.Chrome(chrome_options=option)

    # Define pause time
    pause_time = 3

    # Get url content
    driver.get('https://www.philstar.com/tags/novel-coronavirus')
    print('Getting url content.')

    # Get scroll height
    last_height = driver.execute_script("return document.body.scrollHeight")
    print('Getting scroll height.')

    # Record the starting time
    start = datetime.datetime.now()
    print('Recording start time.')

    while True:
        # Scroll down to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # wait to load page
        time.sleep(pause_time)

        # Calculate new scroll height and compare with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height: # which means end of page
            break
        # update the last height
        last_height = new_height
        ('Updating last height.')
        
    # Record the end time, then calculate and print the total time
    end = datetime.datetime.now()
    delta = end-start
    print("[INFO] Total time taken to scroll till the end {}".format(delta))

    # Extract all article tags
    link_tags = driver.find_elements_by_id('tracking_thumbnail_tag_main')

    u = []
    for tag in link_tags:
        u += tag.find_elements_by_tag_name('a')

    gma_ncov_news = []

    for tag in u:
        gma_ncov_news.append(tag.get_attribute('href'))

    pd.DataFrame(philstar_ncov_news).to_csv('gma.csv', index=False)
```
